```python
import numpy as np
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
X
# X.shape
```

```python
X=X.reshape(-1,1)
X=np.insert(X,0,1,axis=1)
X
```

```python
X.shape
```

```python
def stocastic_gradient_descent(X,y,theta,alpha,epochs):
    m=len(y)
    for _ in range(epochs):
        for _ in range(m):
            idx=np.random.randint(0,m)
            Xi=X[idx]
            yi=y[idx]
            errors=Xi.dot(theta)-yi
            theta0_partial=2*errors
            theta1_partial=2*errors*Xi[1]
            theta[0]-=alpha*theta0_partial
            theta[1]-=alpha*theta1_partial
    return theta
theta=np.zeros(2)
theta=stocastic_gradient_descent(X,y,theta,0.01,100)
r2_score(y_true=y,y_pred=X.dot(theta))
X.dot(theta)


```

```python
r2_score(y_true=y,y_pred=X.dot(theta))
```

```python
from sklearn.linear_model import SGDRegressor
reg=SGDRegressor(learning_rate="constant",eta0=0.01,max_iter=100)
X_hello=X[::,1:]

reg.fit(X_hello,y)

r2_score(y_true=y,y_pred=reg.predict(X_hello))
reg.predict(X_hello)
```
