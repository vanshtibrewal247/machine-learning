```python
import numpy as np
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
X=X.reshape(-1,1)
X=np.insert(X,0,1,axis=1)

```

```python
def mini_batch_gradient(X,y,theta,alpha,epochs,batch_size):
    m=len(y)
    for _ in range(epochs):
        indices= np.random.permutation(m)
        X_suff=X[indices]
        y_suff=y[indices]
        for i in range(0,m,batch_size):
            x_batch=X_suff[i:i+batch_size]
            y_batch=y_suff[i:i+batch_size]
            errors=x_batch.dot(theta)-y_batch
            gradient=2/(len(y_batch))*x_batch.T.dot(errors)
            theta-=alpha*gradient
    return theta
theta=np.zeros(2)
theta=mini_batch_gradient(X,y,theta,0.01,100,10)
r2_score(y_true=y,y_pred=X.dot(theta))

```
