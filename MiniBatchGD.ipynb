```python
import numpy as np
from sklearn.metrics import r2_score,mean_squared_error
```

```python
from sklearn.datasets import load_diabetes
X,y=load_diabetes(return_X_y=True)
X_linear=X.copy()
X=np.insert(X,0,1,axis=1)
```

```python
def mini_batch_gradient_disceint(X,y,theta,alpha,epochs,batch_size):
    m=len(y)
    for _ in range(epochs):
        indices=np.random.permutation(m)
        X_suffled=X[indices]
        Y_suffled=y[indices]
        for i in range(0,m,batch_size):
            X_batch=X_suffled[i:i+batch_size]
            y_batch=Y_suffled[i:i+batch_size]
            errors=X_batch.dot(theta)-y_batch
            gradients=2/(len(y_batch))*(X_batch.T.dot(errors))
            theta-=alpha*gradients
    return theta
theta=np.ones(X.shape[1])
theta=mini_batch_gradient_disceint(X,y,theta,0.1,200,10)



```

```python
r2_score(y_true=y,y_pred=X.dot(theta))
```

```python
mean_squared_error(y_true=y,y_pred=X.dot(theta))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_linear,y)
r2score=r2_score(y_true=y,y_pred=lr.predict(X_linear))
r2score
```

```python
adjusted_r2=1-((1-r2score)*(X_linear.shape[0]-1)/(X_linear.shape[0]-1-X_linear.shape[1]))
adjusted_r2
```
