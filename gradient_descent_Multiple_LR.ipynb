```python
import pandas as pd
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import numpy as np
```

```python
sound=pd.read_csv('AirfoilSelfNoise.csv')
sound
```

```python
X=sound.iloc[:,0:-1]
Y=sound.iloc[:,-1]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
lr=LinearRegression()
lr.fit(X_train,Y_train)
y_predict=lr.predict(X_test)
r2_score(y_pred=y_predict,y_true=Y_test)
```

```python
X_train=np.insert(X_train,0,1,axis=1)
X_test=np.insert(X_test,0,1,axis=1)
```

```python
def gradinet_descent(X,y,alpha,epochs):
    m,n=X.shape
    theta=np.zeros(n)
    for _ in range(epochs):
        y_predict=X.dot(theta)
        errors=y_predict-y
        theta_partial=(2/m)*(X.T.dot(errors))
        theta-=alpha*theta_partial
    return theta
        
```

```python
theta=gradinet_descent(X_train,Y_train,0.02,3000)
y_predictor=X_test.dot(theta)
r2_score(y_true=Y_test,y_pred=y_predictor)

```
